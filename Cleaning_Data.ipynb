```python
import pandas as pd
import numpy as np
description = ["Here are some very simple basic sentences.",
"They won't be very interesting, I'm afraid.", "The point of these examples is to _learn how basic text cleaning works_ on *very simple* data."]
title = ['THRee', 'TYPING IN ALL CAPS!!', 'hello@#$^&*']# change to actual data
df = pd.DataFrame(list(zip(title, description)), columns=['title','description'])
df
```

```python
# concatenate title and product description
df['title_desc'] = df['title'] + ' '+ df['description']
df.head()
```

```python
import nltk
import re
import string as st
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer # or LancasterStemmer, RegexpStemmer, SnowballStemmer
#default_stemmer = PorterStemmer()
default_stopwords = stopwords.words('english') # or any other list of your choice
def clean_text(text):

    def remove_punct(text):
        return ("".join([ch for ch in text if ch not in st.punctuation]))
    
    def tokenize(text):
        text = re.split('\s+' ,text)
        return [x.lower() for x in text]

    def remove_stopwords(text):
        return [word for word in text if word not in nltk.corpus.stopwords.words('english')]

    def stemming(text):
        ps = PorterStemmer()
        return [ps.stem(word) for word in text]

    def lemmatize(text):
        word_net = WordNetLemmatizer()
        return [word_net.lemmatize(word) for word in text]

    text = text.strip(' ') # strip whitespaces
    text = text.lower() # lowercase
    text = remove_punct(text) # remove punctuation
    text = tokenize(text) # tokenize
    text = remove_stopwords(text) # remove stopwords
    text = stemming(text) # stemming
    text = lemmatize(text) # stemming
    
    return text
```

```python
df['clean_text'] = df['title_desc'].apply(lambda x : clean_text(x))
df.head()
```

```python
def return_sentences(tokens):
    return " ".join([word for word in tokens])
```

```python
df['clean_sentences'] = df['clean_text'].apply(lambda x : return_sentences(x))
df.head()
```
